```python
import pandas as pd
df=pd.read_csv('dataset/supermarket_sales.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(data=df,x='Branch',y='Total')
```

```python
sns.countplot(data=df,y='Branch')
```

```python
df
```

```python
fig,axes=plt.subplots(1,2)
sns.barplot(data=df,x='Payment',y='Total',ax=axes[0])
sns.countplot(data=df,x='Branch',ax=axes[1])
plt.show()
```

```python
fig,axes=plt.subplots(2,2)
plt.subplots_adjust(hspace=0.6,wspace=0.4)
sns.barplot(data=df,x='Payment',y='Total',ax=axes[0,0])
sns.countplot(data=df,x='Branch',ax=axes[0,1])
sns.scatterplot(data=df,x='Total',y='Rating',ax=axes[1,0])
sns.lineplot(data=df,x='Product line',y='Total',ax=axes[1,1])
for x in axes.flat:
    plt.setp(x.get_xticklabels(),rotation=90)
    
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(data=df,x='Payment',y='Total',hue='Product line')
```

```python
sns.barplot(data=df,x='Payment',y='Total',hue='Product line')
plt.legend(loc='upper left',bbox_to_anchor=(1.05,1.0))
```

```python
#FacetGrid
pic=sns.FacetGrid(df,col='Branch')
pic.map(sns.barplot,'Product line','Total')
for ax in pic.axes.flat:
    plt.setp(ax.get_xticklabels(),rotation=90)
plt.show()
```

```python
pic=sns.FacetGrid(df,col='Product line',col_wrap=3)
pic.map(sns.barplot,'Branch','Rating')
for ax in pic.axes.flat:
    plt.setp(ax.get_xticklabels(),rotation=90)
plt.show()
```

```python
movies=pd.read_csv('dataset/movie/movies.csv')
movies
```

```python
rating=pd.read_csv('dataset/movie/ratings.csv')
rating
```

```python
#merge, join

merge_file=pd.merge(rating,movies,how='inner',indicator=True)
merge_file
```

```python
merge_file=pd.merge(rating,movies,how='outer',indicator=True)
merge_file
```

```python
merge_file=pd.merge(rating,movies,how='left',indicator=True)
merge_file
```

```python
merge_file=pd.merge(rating,movies,how='right',indicator=True)
merge_file
```

```python
movies.join(rating,rsuffix='outer')
```

```python
tags=pd.read_csv('dataset/movie/tags.csv')
tags
```

```python
alldf=movies.join(rating,rsuffix='inner').join(tags,rsuffix='inner')
alldf

#OR

"""two_df=movies.join(rating,rsuffix='inner')
all_df=two_df.join(tags,rsuffix='inner')"""
```

```python
alldf.groupby('title')['rating'].mean()[:10].plot(kind='bar')
```

```python
rating_value=alldf.groupby('title')['rating'].mean()

myvideo=[]
for x in rating_value:
    if x>3.5:
        myvideo.append(rating_value)
```
