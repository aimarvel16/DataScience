```python
def time_fun(x):
    if x<=4:
        return 'Early Morning'
    elif x>4 and x<12:
        return 'Morning'
    elif x>=2 and x<18:
        return 'Afternoon'
    elif x>=18 and x<=20:
        return 'Evening'
    else:
        return 'Night'
    

df['Time_slot']=df['Hour'].apply(time_fun)
```
