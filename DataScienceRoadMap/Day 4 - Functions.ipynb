```python
import pandas as pd
df=pd.read_csv('dataset/Pokemon.csv')
df
```

```python
def myfunc(a,b):
    for x,y in zip(a,b):
        value= (x+y)/2
        if value>100:
            return 1
        else:
            return 0
    
df['Attack'].fillna(myfunc(df['Total'],df['Speed']))
```

```python
# fill na of some column with three colums's mean value , but mean value>200 True ,else False

def calculate(a,b,c):
    for x,y,z in zip(a,b,c):
        value=(x+y+z)/3
        if value>200:
            return True
        else:
            return False

df['Defense'].fillna(calculate(df['Total'],df['HP'],df['Attack']))
```
