```python
import pandas as pd
df=pd.read_csv('dataset/Pokemon.csv')
df
```

```python
def myfun(df):
    r_value={}
    for index,row in df.iterrows():
        value=(row['HP']+row['Defense'])/2
        if value>80:
             r_value[index]=1
        else:
            r_value[index]=0
    return r_value

df['Attack'].fillna(myfun(df[['HP','Defense']]))
```
