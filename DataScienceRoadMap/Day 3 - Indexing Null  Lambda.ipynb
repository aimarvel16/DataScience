```python
def myfun(data1):
    for x in data1:
        if x>3:
            return 0
        else:
            return 1
    

df['HP'].fillna(myfun(df['Generation']))
```

```python
def myfun(d1,d2,d3):
    value= (d1*d2*d3)/2
    return value
df['HP'].fillna(myfun(df['Total'],df['Attack'],df['Defense']))
```

```python
def myfun(data1,data2):
    for x,y in zip(data1,data2):
        value=x*y
        if value>100:
            return 0
        else:
            return 1
    
df['HP'].fillna(myfun(df['Generation'],df['Attack']))
```
